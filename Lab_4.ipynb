В данных есть проблема: 
таблица слишком большая и в ней объединены как личные данные клиента, так и данные о его покупках.Надо разделить

```python
#Данные по самому клиенту
data_1 = data.iloc [:, 0:8]
data_1["Complain"] = data["Complain"]
data_1
```

```python
#Данные о покупках клиента
data_2 = data.iloc [:, 8:14]
data_2["ID"] = data["ID"]
data_2 = data_2.reindex(columns=['ID','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds'])
data_2=data_2.rename(columns={"MntWines": "Wines", "MntFruits": "Fruits","MntMeatProducts": "Meat",'MntSweetProducts':'Sweet',"MntFishProducts": "Fish","MntGoldProds": "Gold"})
data_2
```

В таблице выше названия столбцов также являются значениями, поэтому преобразуем следующим образом:

```python
data_2.melt(id_vars=['ID'],
                 value_vars=["Wines", "Fruits","Meat",'Sweet', "Fish","Gold"],
                 var_name='Product',
                 value_name='Quantity').sort_values(by=['ID'])
```

Если нужно показать зависимость, то можно сгруппировать данные.

```python
#Покажем как изменяется сумма покупок в зависимости от образования и семейного положения

#удалим слобцы, которые не несут важной для статистики информации
df=data.drop(['Year_Birth','Dt_Customer','ID'],axis=1)

#объединим малышей и подростков в одну группу
df['Kids_all'] = df['Kidhome']+df['Teenhome']
df=df.drop(['Kidhome','Teenhome'],axis=1)

d = df.groupby(['Education','Marital_Status'], as_index = False).sum().set_index(['Education','Marital_Status'])
d.head(15)
```

Так как таблица довольно большая можно  применить принцип разделяй и властвуй, то есть сделать некую декомпозицию основной таблицы, чтобы показать частные случаи

```python
#Например, зависимость суммы на покупку фруктов от количества детей
f=['MntFruits']
df3 = df.groupby(['Kids_all'])[f].agg(['mean', 'max', 'min']).astype(int)
df3
```

```python
f=['MntSweetProducts']
df4 = data.groupby(['Kidhome', 'Teenhome'])[f].agg(['mean', 'max', 'min']).astype(int)
df4
```

```python
#Или соотношение между суммой потраченной на покупку продуктов и размером дохода у людей с разным образованием
df['Sum_all_prod']=df.eval('MntWines+MntFruits+MntMeatProducts+MntSweetProducts+MntFishProducts+MntGoldProds')
f=['Income','Sum_all_prod']
df.groupby('Education')[f].mean().astype(int).sort_values(by=['Income'])
```
